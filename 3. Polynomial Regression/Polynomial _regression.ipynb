# Data Preprocessing

## Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Import the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

# Traning Model

## Training the Linear Regression

```python
from sklearn.linear_model import LinearRegression
li_reg = LinearRegression()
li_reg.fit(X, y)
```

## Training the Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=4)
X_poly = pf.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
```

# Visualizing

## Visualize the Linear Regression results

```python
plt.scatter(X, y, color='red')
plt.plot(X, li_reg.predict(X), color='blue')
plt.title('Linear Regression')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

## Visualize the Polynomial Regression results

```python
plt.scatter(X, y, color='red')
plt.plot(X, poly_reg.predict(X_poly))
plt.title('Linear Regression')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

# Predict with Linear Regression

```python
print(li_reg.predict([[6.5]]))
```

# Predict results with Polynomial Regression

```python
print(poly_reg.predict(pf.fit_transform([[6.5]])))
```
